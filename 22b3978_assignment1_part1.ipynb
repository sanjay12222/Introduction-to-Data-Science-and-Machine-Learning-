#2.Import Neccessary Libraries

```python
# Importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

#3.Exploratory Data Analysis

```python
# Checking the data types of both datasets
print(restaurant_df.dtypes)
print(housing_df.dtypes)
```

```python
# Checking for missing values in both datasets
print(restaurant_df.isnull().sum())
print(housing_df.isnull().sum())


```

```python
#Drop missing values
restaurant_df.dropna(inplace=True)
housing_df.dropna(inplace=True)
```

```python
# Extracting ZIP codes (assuming they exist in columns related to location/address)
housing_df['ZIP'] = housing_df['SITE ADDRESS'].str.extract(r'(\d{5})')  # Extract 5-digit ZIP codes from addresses
restaurant_df['ZIP'] = restaurant_df['facility_zip'].astype(str).str[:5]  # Using first 5 digits for uniformity


```

```python
# Checking for unexpected entries (invalid ZIPs)
print(housing_df['ZIP'].unique())
print(restaurant_df['ZIP'].unique())
```

#4.Visualise Data

```python
# Plotting a distribution of health inspection scores
plt.figure(figsize=(10,6))
sns.histplot(restaurant_df['score'], bins=20, kde=True)
plt.title("Distribution of Restaurant Health Scores")
plt.show()

```

```python
# Visualizing number of affordable housing projects by ZIP code
plt.figure(figsize=(10,6))
housing_df['ZIP'].value_counts().plot(kind='bar')
plt.title("Number of Affordable Housing Projects by ZIP Code")
plt.show()
```

#5.Summarizing Data by ZIP Code Using SQL:

```python
# Summarizing restaurant data by ZIP code
restaurant_summary = restaurant_df.groupby('ZIP').agg({'score': ['mean', 'count']}).reset_index()
restaurant_summary.columns = ['ZIP', 'Mean_Score', 'Restaurant_Count']



# Displaying the summaries
restaurant_summary.head()
```

```python
# Summarizing housing data by ZIP code
housing_summary = housing_df.groupby('ZIP').size().reset_index(name='Housing_Project_Count')
#display the summaries
housing_summary.head()
```

#6.Joining Data by ZIP Code Using SQL:

```python
# Merging the two datasets by ZIP code
merged_df = pd.merge(restaurant_summary, housing_summary, on='ZIP', how='inner')

# Display the merged data
merged_df.head()
```

#7.Formulating and Testing the Hypothesis

```python
# Calculating Pearson correlation between the number of housing projects and restaurant scores
corr = merged_df[['Mean_Score', 'Housing_Project_Count']].corr()

# Displaying correlation result
print(corr)
```

```python

# Visualization of the relationship
plt.figure(figsize=(8,6))
sns.regplot(x='Housing_Project_Count', y='Mean_Score', data=merged_df)
plt.title('Relationship between Housing Projects and Health Scores by ZIP Code')
plt.show()
```
